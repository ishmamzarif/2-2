## scatterplots and lineplots

```python
X = np.random.random(50) * 50 # 50 random points from 0 to 50
Y = np.random.random(50) * 2500 # 50 random points from 0 to 2500


x = [0] * 50
for i in range(50):
  x[i] = i

x = np.array(x)
y = x * x


# scatter just plots the points
plt.scatter(X, Y, color = "red", s = 20, marker = ".")

#plt.plot joins the points
plt.plot(x, y, color = "green", linestyle = "solid")

# color
# s for size
# marker (for scatter)
# linewidth (for plot)
# linestyle (for plot)

plt.show()
```

### common parameters

```python
x = np.linspace(-2 * np.pi, 3 * np.pi, 100) # 100 points between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
plt.plot(x, y1, label='sine wave', color='blue', linestyle='dotted')
plt.plot(x, y2, label='cosine Wave', color='red', linestyle='--')

# Add labels and a title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Sine and Cosine Waves')

# add a legend to differentiate the lines
# useful if there are multiple lines
plt.legend()

# Add a grid for better readability
plt.grid(False)

# axvline -> add vertical line x = 2
# axhline -> add horizontal line y = 0
plt.axvline(0, color='black', linewidth=0.8)
plt.axhline(0, color='black', linewidth=0.8)

# Display the plot
plt.show()
```

### subplots

```python
x = np.arange(100)

# one figure with 4 grids
fig, axs = plt.subplots(2, 2)


axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("sine")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("cosine")

axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title("tan")

axs[1, 1].plot(x, np.exp(x))
axs[1, 1].set_title("exp")

plt.show()
```

## plot surfaces

```python
ax = plt.axes(projection = "3d")

x  = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap = "Spectral")

plt.show()
```

## animations

```python
import random
from IPython.display import clear_output
```

```python
heads_tails = [0, 0]

# Create figure ONCE
fig, ax = plt.subplots()
bars = ax.bar(["heads", "tails"], heads_tails, color=["r", "b"])
ax.set_ylim(0, 1000)

# Display once and keep updating it
display(fig)

for i in range(1000):
    heads_tails[random.randint(0, 1)] += 1

    # Update bar heights
    bars[0].set_height(heads_tails[0])
    bars[1].set_height(heads_tails[1])

    # Update title
    ax.set_title(f"Heads={heads_tails[0]}, Tails={heads_tails[1]} (Iteration: {i+1})")

    # Redraw canvas
    fig.canvas.draw()

    # Replace output smoothly
    clear_output(wait=True)
    display(fig)

plt.close(fig)
```
