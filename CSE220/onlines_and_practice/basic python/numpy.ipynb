```python
class Employee:
    class_name = 'Employee'

    # double underscore -> dunder
    def __init__(self, id, name):
        self.id = id
        self.name = name

    def __str__(self):
        return f"{self.id} with name {self.name}"
    
    def __repr__(self):
        return f"{self.name} has the id {self.id}"

employee = Employee(2596, "Kirk")
print(employee)
employee
```
