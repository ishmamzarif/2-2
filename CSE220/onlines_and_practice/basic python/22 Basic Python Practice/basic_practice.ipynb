```python
### 3 ###

class Student:
    def __init__(self, name, std_id, grades):
        self.name = name
        self.std_id = std_id
        self.grades = grades

    def __str__(self):
        return f"Student, Name: {self.name}, ID: {self.std_id}"

    def add_grade(self, grade):
        self.grades.append(grade)

    def get_average(self):
        n = len(self.grades)
        total = sum(self.grades)
        return total / n
        
    def get_letter_grade(self):
        avg = self.get_average()
        
        if avg > 80:
            return "A+"
        elif avg > 75:
            return "A"
        elif avg > 70:
            return "B"
        elif avg > 60:
            return "C"
        elif avg > 50:
            return "D"
        else:
            return "F"


s1 = Student("Zarif", 35, [80, 90, 95])
s2 = Student("Mehwish", 95, [70, 81, 55])
s3 = Student("Karim", 9, [40, 63, 77])

s1.add_grade(99)
print(s1.get_average())
print(s1.get_letter_grade())

print(s2)
print(s2.get_average())
print(s2.get_letter_grade())

s3.add_grade(22)
print(s3)
print(s3.get_average())
print(s3.get_letter_grade())
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
product_a = np.array([23, 25, 28, 30, 35, 38, 42, 45, 48, 50, 53, 55])
product_b = np.array([30, 32, 31, 33, 35, 37, 39, 41, 43, 45, 47, 50])
product_c = np.array([15, 18, 22, 25, 28, 32, 35, 38, 42, 45, 48, 52])

fig, axes = plt.subplots(2, 2)
fig.set_size_inches(12, 8)

axes[0, 0].plot(months, product_a, color = "blue", linestyle = "solid")
axes[0, 0].plot(months, product_b, color = "green", linestyle = "dashed")
axes[0, 0].plot(months, product_c, color = "red", linestyle = "dotted")
axes[0, 0].legend()
axes[0, 0].grid(True)
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Sales (In thousands)')
axes[0, 0].set_title('Monthly Sales Trend 2024')

axes[0, 1].bar(['Product A', 'Product B', 'Product C'],
               [product_a[11], product_b[11], product_c[11]],
               color = ["blue", "green", "red"])
axes[0, 1].yaxis.grid(True)
axes[0, 1].set_xlabel('Product')
axes[0, 1].set_ylabel('Sales (In thousands)')
axes[0, 1].set_title('December Sales Comparison')


axes[1, 0].scatter(months, product_a, color = "blue")
axes[1, 0].grid(True)
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Sales (In thousands)')
axes[1, 0].set_title('Product A Sales Distribution')


p1 = np.array([sum(product_a[0:3:1]), sum(product_a[3:6:1]), sum(product_a[6:9:1]), sum(product_a[9:12:1])])
p2 = np.array([sum(product_b[0:3:1]), sum(product_b[3:6:1]), sum(product_b[6:9:1]), sum(product_b[9:12:1])])
p3 = np.array([sum(product_c[0:3:1]), sum(product_c[3:6:1]), sum(product_c[6:9:1]), sum(product_c[9:12:1])])


axes[1, 1].bar(["01", "02", "03", "04"], 
               p1,
               color = "blue")
axes[1, 1].bar(["01", "02", "03", "04"], 
                p2,
               color = "green", 
               bottom = p1)
axes[1, 1].bar(["01", "02", "03", "04"], 
                p3,
               color = "red", 
               bottom = p1 + p2)
axes[1, 1].legend()
axes[1, 1].legend()
axes[1, 1].yaxis.grid(True)

fig.tight_layout()
fig.suptitle('2024 Product Sales Analysis')
```
