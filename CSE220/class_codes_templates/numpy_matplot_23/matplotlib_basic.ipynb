```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 100)
y = np.random.random(99) * 100

plt.plot(x, y, 
         marker = 'o', 
         linestyle = 'dashed', 
         alpha = 0.5, 
         color = "purple")
```

```python
plt.subplot(2, 1, 1)             # (rows, cols, index) -> select 1st cell
plt.plot(x, np.sin(x), color="steelblue")
plt.title("sin(x)")              # title for THIS subplot only
plt.ylabel("amplitude")
plt.axhline(y=0, color="black", linewidth=0.8)  # horizontal ref line at y=0
 
# --- Bottom plot: cell index 2 ---
plt.subplot(2, 1, 2)             # select 2nd cell (below the first)
plt.plot(x, np.cos(x), color="tomato")
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("amplitude")
plt.axhline(y=0, color="black", linewidth=0.8)
 
plt.tight_layout()   # auto-fixes overlapping titles/labels between subplots
plt.suptitle("plt.subplot() demo", y=1.02, fontsize=14, fontweight="bold")
plt.show()
```

```python
# --- left plot: vertical bar ---
plt.subplot(1, 2, 1)        # 1 row, 2 cols, 1st cell
plt.bar(x, np.random.random(len(x)) * 100, color = ["steelblue"])
plt.title("Vertical Bar")

# --- right plot: horizontal bar ---
plt.subplot(1, 2, 2)        # 1 row, 2 cols, 2nd cell
plt.barh(x, np.random.random(len(x)) * 50, color = ["tomato"])
plt.title("Horizontal Bar")

plt.tight_layout()
plt.show()
```

```python
# np.sign()
# plt.stem()


arr = np.array([-5, 0, 3, -2, 7, 0])

result = np.sign(arr)
print(result)
# [-1  0  1 -1  1  0]

# np.sign() returns:
#  -1  if the number is negative
#   0  if the number is zero
#   1  if the number is positive
# Works element-wise on the whole array -- no loop needed.

# Common use case: get the direction/polarity of values
prices_change = np.array([2.5, -1.3, 0, 4.1, -0.5])
direction = np.sign(prices_change)
print(direction)
# [ 1. -1.  0.  1. -1.]   --> up, down, unchanged, up, down

# Plotting original values vs their sign
n = np.arange(len(prices_change))

fig, ax = plt.subplots(figsize=(7, 4))
ax.stem(n, prices_change, linefmt="gray", markerfmt="ko", basefmt=" ", label="original")
ax.stem(n, direction, linefmt="r-", markerfmt="r^", basefmt=" ", label="sign")
# markerfmt -> controls the tip (dot o, diamond d, triangle ^, square s)
# basefmt -> controls the horizontal basline at y = 0
# linefmt -> controls the line format for each line


ax.axhline(y=0, color="black", linewidth=0.8)
ax.set_title("np.sign() demo")
ax.legend()
plt.show()
```

```python
# square wave in matplotlib

t = np.linspace(0, 4, 1000)                   # time axis
square_wave = np.sign(np.sin(2 * np.pi * t)) * 5 # +5 or -5 depending on sign of sin

plt.plot(t, square_wave)

plt.title("Square Wave")
plt.xlabel("time")
plt.ylabel("amplitude")

plt.axhline(y=0, color="black", linewidth=0.8)
plt.show()
```
